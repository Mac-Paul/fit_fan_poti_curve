```python
from IPython.display import display, Markdown, SVG
import numpy as np
import pandas as pd
import altair as alt
from scipy.optimize import curve_fit
data = pd.read_csv("angles-values.csv", names=["Winkel", "Spannung"], index_col=False, header=0)
plot = alt.Chart(data).mark_line(point=False).encode(
        alt.X("Spannung", title="Spannung"),
        alt.Y("Winkel:Q", title="Winkel")).properties(width=400, height=400)
plot
```

```python
def f(x, a, b, x_off, y_off):
    s1 = a*pow(x-x_off, 33)
    return s1+b*x+y_off

data = pd.read_csv("angles-values.csv", names=["Winkel", "Spannung"], index_col=False, header=0)
angles = np.array(data["Winkel"])
voltages = np.array(data["Spannung"])
voltages
params, _ = curve_fit(f, xdata=voltages, ydata=angles, p0=[3e-86, 5.3e-3, 462, 181], maxfev=10000)
print(params)
#params=[3e-65, 5e-3, 475, 181]
data["Kurve"] = "gemessen"
simulation = []
for v in data["Spannung"]:
    simulation.append(np.array([f(v, *params), v, "berechnet"]))
simulation = pd.DataFrame(simulation, columns=["Winkel", "Spannung", "Kurve"])
data = simulation.append(data, ignore_index=True)
plot1 = alt.Chart(data).mark_line(point=False).encode(
    alt.X("Spannung", title="Spannung"),
    alt.Y("Winkel", title="Winkel"),
    color=alt.Color("Kurve:N")
).properties(width=400, height=400)
plot2 = alt.Chart(data).mark_line(point=True).encode(
        alt.X("Winkel:Q", title="Winkel in Grad"),
        alt.Y("Spannung:Q", title="Spannung in 1/1024 V", 
              scale=alt.Scale(type="log", base=10, domain=[8, 890])),
        color=alt.Color("Kurve:N", 
                        scale=alt.Scale(scheme="dark2"),
                        legend=alt.Legend(
            orient="none", legendX=20, legendY=20,
            fillColor="white", strokeColor="black", cornerRadius=7, padding=6,
            title="Kurven"))
).properties(width=800, height=300)
plot2
```
